# WebScraper Bot

Extracts text from 'horoscope.com' and outputs a json file with 2 months worth of horoscopes (for each sign)

```python
import pandas as pd
from datetime import date, timedelta
from dateutil.relativedelta import relativedelta

current_date = date.today()

startDate = date(2022,1,1)
endDate = current_date

datesIn2021 = pd.date_range(startDate,endDate,freq='d')

# dates = list(datesIn2021.strftime('%Y%m%d').str.upper())
datelist = list(datesIn2021)
dates = [x.date() for x in datelist]
print(len(dates))
```

```python
# checking before scraping
from bs4 import BeautifulSoup
import json
import requests
from lxml.html import fromstring
import re
import string, os

for sign in range(1,2):
    
    result = []
    
    for date in dates[:3]:
        strDate = date.strftime('%Y%m%d').upper()
        URL = "https://www.horoscope.com/us/horoscopes/general/horoscope-archive.aspx?sign=" + str(sign) + "&laDate=" + strDate
        page = requests.get(URL)
        soup = BeautifulSoup(page.text, "lxml")

        horoscope = soup.find("div", attrs={"class":"main-horoscope"})
        horoscope = horoscope.find("p")
        text = horoscope.text
        
        # clean the data and remove date at the start
        out = re.sub('^.+-\s', '', text)
        out = "".join(v for v in out if v not in string.punctuation).lower()
        out = out.lower()
        
        result.append({"date": date.strftime("%Y-%m-%d"), "text": out,})
        
    print(result)

```
